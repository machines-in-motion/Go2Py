# Lowlevel Control

```python
from Go2Py.robot.interface import GO2Real
import time
robot = GO2Real(mode='lowlevel')
```

```python
robot.getJointStates()
```

```python
import numpy as np
import time
for i in range(1000000):
    q = np.zeros(12) 
    dq = np.zeros(12)
    kp = np.ones(12)*0.0
    kd = np.ones(12)*0.0
    tau = np.zeros(12)
    tau[0] = 0.0
    robot.setCommands(q, dq, kp, kd, tau)
    time.sleep(0.01)    
```
